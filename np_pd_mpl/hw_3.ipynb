## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('D:\\Datasets\\ml-latest-small\\ratings.csv')
ratings.head()
```

```python
def classification(movie_rating):
    """Функция классифицирует фильмы по категориям в зависимости от рейтинга."""
    
    if movie_rating <= 2.0:
        return 'low rating'
    elif movie_rating <= 4.0:
        return 'middle rating'
    # непонятно, почему по условиям задачи не нужно обрабатывать случаи 4 < rating < 4.5 и 4.5 < rating < 5, но ладно...
    elif movie_rating == 4.5 or movie_rating == 5.0:
        return 'high rating'
```

```python
ratings['class'] = ratings['rating'].apply(classification)
ratings.head()
```

## Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

Результат классификации запишите в отдельный столбец region.

```python
df = pd.read_csv('D:\\Datasets\\ml-latest-small\\keywords.csv')
df.head()
```

```python
def geo_classification(query):
    """Функция классифицирует поисковые запросы по регионам."""
    
    for region, city in geo_data.items():
        for word in query.split():
            if word in city:
                return region
    return 'undefined'
```

```python
df['region'] = df['keyword'].apply(geo_classification)
df.head()
```

```python
df[df['region'] != 'undefined'].head()
```

## Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из [базы](https://grouplens.org/datasets/movielens)). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

1. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

    * для каждой строки пройдите по всем годам списка years
    * если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    * если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
1. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

1. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
movies = pd.read_csv('D:\\Datasets\\ml-latest-small\\movies.csv')
movies.head()
```

```python
# 1. В переменную years запишите список из всех годов с 1950 по 2010.
years = list(range(1950, 2011))
print(years)
```

```python
# 2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска.
def production_year(film):
    """Функция возвращает год из названия фильма"""
    
    for year in years:
        if str(year) in film:
            return year
    return 1900
```

```python
# 3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
movies['year'] = movies['title'].apply(production_year)
movies.head()
```

```python
# 4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

avg_rating = ratings['rating'].groupby(ratings['movieId']).mean()
avg_rating
```

```python
movies = movies.set_index('movieId').join(avg_rating, how='left', on='movieId')
movies.head()
```

```python
movies['rating'].groupby(movies['year']).mean().sort_values(ascending=False).head(10)
```
